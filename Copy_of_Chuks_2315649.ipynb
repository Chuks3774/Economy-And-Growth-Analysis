# **4**.**Data_Analysis**

# **4.1** **GDP** **per** **capita** **growth** **rate** **of** **20** **countries**

To find the GDP  per capital growth rate of 20 various countries,we check for the unique values in the year column(year 2023) and filter out,then we sort in descending other and select top 20.We chose to investigate top 20 countries with highest GDP per capital in the most current year, because we are interested in countries with high economic stability for investment or business expansion.





```python
# Check unique values in the 'Year' column
print(Merged_Ecoperpop['Year'].unique())

# Convert 'Year' column to integer
Merged_Ecoperpop['Year'] = Merged_Ecoperpop['Year'].astype(int)

# Filter the data for 2023
Merged_Ecoperpop_2023 = Merged_Ecoperpop.loc[Merged_Ecoperpop['Year'] == 2023]

# Display the filtered data
print(Merged_Ecoperpop_2023)
```

we have 153 rows and 4 columns,we sort the dataset by GDP per capita in descending order and select the top 20 countries

```python
# Sort by GDP per capita and select the top 20
Merged_Ecoperpop_2023 = Merged_Ecoperpop_2023.sort_values(by='gdp_per_capita', ascending=False)
Merged_Ecoperpop_2023 = Merged_Ecoperpop_2023.head(20)
Merged_Ecoperpop_2023
```

since we want to visualize the top 20 countries,we create a new dataframe to target the data and set year as index for time series

```python
# create a new dataframe in a variable highest_gdp_percapita which countain gdp_per_capita only with year as the index
highest_gdp_percapita = Merged_Ecoperpop_2023.set_index('Year')['gdp_per_capita']
highest_gdp_percapita
```

since the values are large,we scale the data by dividing with 1000 and round in 2 decimal places.

```python
# Divide GDP per capita by one thousand and round to 2 decimal places
highest_gdp_percapita = (highest_gdp_percapita / 1000).round(2)
highest_gdp_percapita
```

Visualize the top 20 countries with the highest gdp per capita with a bar chart to see the distribution

```python
# Plot a bar chart
import plotly.express as px
fig = px.bar(Merged_Ecoperpop_2023, x='Country Name', y='gdp_per_capita', title='Top 20 Countries with the Highest GDP per capita')
fig.show()

```

Among the 20 nations investigated,luxembourg remain the country with the highest GDP per capita while New zealand the lowest.It can also be seen that most European nations are economically stable.

#**4.2** **Finding** **the** **countries** **with** **the** **highest** **and** **lowest** **GDP** **per** **capita**

From the table above,we can compute the country with the highest GDP per capita growth and that of the lowest growth.

This result wil help investors  know the country(countries) with high GDP growth and that with  lowest growth rate.This is necessary to inform their business decision on where to invest or expand

```python
# Find country with highest and lowest gdp per capita only
highest_gdp_percapcountry = Merged_Ecoperpop_2023.loc[Merged_Ecoperpop_2023['gdp_per_capita'].idxmax()]['Country Name']
lowest_gdp_percapcountry = Merged_Ecoperpop_2023.loc[Merged_Ecoperpop_2023['gdp_per_capita'].idxmin()]['Country Name']

# These variables where printed for easy read
print(f"Country with highest GDP per capita growth: {highest_gdp_percapcountry}")


print(f"Country with lowest GDP per capita growth: {lowest_gdp_percapcountry}")
```

Luxembourg is the country with the highest Gdp per capital while New zealand is with the lowest per capita amoung the top 20 countries investigated.

# **4.3** **Finding** **GDP** **per**  **capita** **growth** **trend** **over** **time** **for** **luxembourg**

Since luxembourg is the country with the highest gdp per capita,we explore its growth rate to know how consistent it is over the years.This will aim investor on how Gdp per capita consistently grow  over time

*   filter for luxembourg from the merged dataframe
*   extract the last 10 years for luxembourg and plot the line chart to visualise



```python
# Filter for Luxembourg data and set index to 'Year'
luxembourg_data = Merged_Ecoperpop[Merged_Ecoperpop['Country Name'].str.lower() == 'luxembourg'].set_index('Year')

# Get the last 10 years of data
last_10_years = luxembourg_data.tail(10)

# Plot a line chart for the last 10 years
fig = px.line(last_10_years, x=last_10_years.index, y='gdp_per_capita', title='Luxembourg GDP per capita over the last 10 years')
fig.show()
```

Luxembourg's swift recovery from economic decline in 2022 indicates that its policies foster a stable and resilient economy.They demonstrate this resilience and adaptability in economic trends.

# **4.4** **Finding** **the** **relationship** **between** **GDP** **per** **capita**  **and** **population**

To understand some economic factors affect gdp per capita,we compare the relationship between gdp per capita and population(economic factor).This will help investors to know how this key indicator of economic performance affects the average living standards most espercially in luxembourg,country in view.

*   we  visualize this relationship.




```python
# create a scatter plot to analyses the relationship between GDP per capita and population
fig = px.scatter(Merged_Ecoperpop_2023
                 , x='gdp_per_capita'
                 , y='Population'
                 , color='Country Name'
                 , title='Relationship between GDP per capita and Population'
                 , labels={'gdp_per_capita': 'GDP per capita', 'Population': 'Population'})
fig.show()


```

The dataset shows a poor negative correlation between population total and GDP per capita, as GDP per capita decreases population tends to increases.However,this is not the case in all circumstances,for instance,United states though the population is high,the gdp per capital is moderately high too.The outlier shows that united state have the higest population among the investigated countries. It is visibly clean the luxembourg is a country to invest.
